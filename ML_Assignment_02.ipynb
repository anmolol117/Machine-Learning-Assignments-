```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

df_new['BirthDate'] = pd.to_datetime(df_new['BirthDate'], errors='coerce') 
today = pd.Timestamp.today()
df_new['Age'] = (today - df_new['BirthDate']).dt.days // 365   
df_new = df_new.drop(columns=['BirthDate'])  
```

```python
categorical_cols = ['City', 'StateProvinceName', 'CountryRegionName',
                    'Education', 'Occupation', 'Gender', 'MaritalStatus']
numeric_cols = ['NumberCarsOwned', 'NumberChildrenAtHome',
                'TotalChildren', 'YearlyIncome', 'AvgMonthSpend', 'Age']
target = 'BikeBuyer'
```

```python
categorical_pipeline = Pipeline(steps=[ ("imputer", SimpleImputer(strategy="most_frequent")), 
                                         ("encoder", OneHotEncoder(handle_unknown="ignore")) ])

numeric_pipeline = Pipeline(steps=[ ("imputer", SimpleImputer(strategy="median")), 
                                      ("scaler", StandardScaler()) ])

preprocessor = ColumnTransformer(
    transformers=[
        ("categorical", categorical_pipeline, categorical_cols),
        ("numeric", numeric_pipeline, numeric_cols)
    ])

X = df_new.drop(columns=[target])
y = df_new[target]

X_processed = preprocessor.fit_transform(X)

print("Shape before:", X.shape)
print("Shape after:", X_processed.shape)
```

Make sure each attribute is transformed in a same scale for numeric attributes and Binarization for each

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, StandardScaler

# --- Example raw data ---
data = pd.DataFrame({
    "CommuteDistance": [5, 10],
    "YearlyIncome": [60000, 90000],
    "Education": ["Graduate", "PhD"],
    "BikeBuyer": [1, 0]
})

# --- Step 1: Categorical (one-hot) ---
categorical_cols = ["Education", "BikeBuyer"]
numeric_cols = ["CommuteDistance", "YearlyIncome"]

encoder = OneHotEncoder(handle_unknown="ignore", sparse_output=False)
cat_transformed = encoder.fit_transform(data[categorical_cols])

# --- Step 2: Numeric (standardized) ---
scaler = StandardScaler()
num_transformed = scaler.fit_transform(data[numeric_cols])

# --- Step 3: Combine ---
transformed = np.hstack([num_transformed, cat_transformed])
df_transformed = pd.DataFrame(transformed)
print("Transformed Data:\n", df_transformed)
```

```python
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics.pairwise import cosine_similarity

encoder = OneHotEncoder(sparse_output=True)  
encoded = encoder.fit_transform(df_new[['YearlyIncome', 'BikeBuyer']])  

x = encoded[0].toarray().flatten()   
y = encoded[1].toarray().flatten()


def simple_matching(a, b):
    return np.mean(a == b)

def jaccard_similarity(a, b):
    intersection = np.sum(np.logical_and(a == 1, b == 1))
    union = np.sum(np.logical_or(a == 1, b == 1))
    return intersection / union if union != 0 else 0

cos_sim = cosine_similarity([x], [y])[0,0]

print("Simple Matching:", simple_matching(x, y))
print("Jaccard:", jaccard_similarity(x, y))
print("Cosine Similarity:", cos_sim)

corr = df_new['BikeBuyer'].corr(df_new['YearlyIncome'])
print("Correlation (Commute Distance vs Yearly Income):", corr)
```
